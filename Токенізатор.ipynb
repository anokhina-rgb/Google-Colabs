```python
import requests
from bs4 import BeautifulSoup
import spacy

# 1. CRAWLER: Отримуємо текст із веб-сторінки
url = "https://example.com/ukrainian-article" # Приклад посилання
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
raw_text = soup.get_text()

# 2. TAGGER: Проводимо частиномовну розмітку (напр. українська мова)
nlp_uk = spacy.load("uk_core_news_trf")
doc = nlp_uk(raw_text[:500]) # беремо перші 500 символів для прикладу

# 3. РЕЗУЛЬТАТ: Таблиця з розміткою
print(f"{'Слово':<15} | {'Частина мови (POS)':<20}")
print("-" * 40)
for token in doc:
    if not token.is_space and not token.is_punct:
        print(f"{token.text:<15} | {token.pos_:<20}")
```
