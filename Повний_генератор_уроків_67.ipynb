```
import sys
import subprocess
import io
import re
import os
import base64
from collections import Counter
from IPython.display import display, Javascript

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

try:
    from docx import Document
    from docx.shared import Pt, Inches
    from googletrans import Translator
except ImportError:
    install("python-docx")
    install("googletrans==4.0.0-rc1")
    from docx import Document
    from docx.shared import Pt, Inches
    from googletrans import Translator

import ipywidgets as widgets

translator = Translator()

def clean_text(text):
    # Агресивна очистка: видаляємо всі технічні блоки
    text = re.sub(r'[-]{10,}.*?[-]{10,}', '', text)
    text = re.sub(r'Page \d+', '', text)
    text = re.sub(r'Contents', '', text, flags=re.IGNORECASE)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

def get_translation(text, dest='uk'):
    try:
        return translator.translate(text, dest=dest).text
    except:
        return "[Помилка перекладу]"

def auto_download_js(filename):
    with open(filename, "rb") as f:
        bytes_content = f.read()
    b64 = base64.b64encode(bytes_content).decode()
    js = f'''
    var link = document.createElement('a');
    link.href = 'data:application/vnd.openxmlformats-officedocument.wordprocessingml.document;base64,{b64}';
    link.download = '{filename}';
    link.click();
    '''
    display(Javascript(js))

def process_full_book(file_content):
    doc = Document(io.BytesIO(file_content))
    # Збираємо чистий текст
    raw_text = " ".join([clean_text(p.text) for p in doc.paragraphs if len(p.text.strip()) > 10])
    sentences = re.split(r'(?<=[.!?])\s+', raw_text)

    # Ділимо на уроки по 20 речень
    lesson_size = 20
    lesson_num = 1

    for i in range(0, len(sentences), lesson_size):
        lesson_sentences = sentences[i:i + lesson_size]
        if len(lesson_sentences) < 5: break # Ігноруємо залишки

        text_block = " ".join(lesson_sentences)

        # 15 багатокомпонентних термінів (2-4 слова)
        term_pattern = r'\b(?:[A-Z][a-z]+(?:\s+[A-Z][a-z]+){1,3})\b'
        terms = re.findall(term_pattern, text_block)
        unique_terms = [t for t in dict.fromkeys(terms) if len(t.split()) > 1][:15]

        # Генерація файлу
        filename = f"Урок_{lesson_num}.docx"
        new_doc = Document()
        section = new_doc.sections[0]
        section.page_width = Inches(8.27)
        section.page_height = Inches(11.69)

        new_doc.add_heading(f'Урок {lesson_num}', 0)
        new_doc.add_heading('Original Text (ENG)', level=1)
        new_doc.add_paragraph(text_block)

        new_doc.add_heading('Key Terms & Phrases (15)', level=1)
        table = new_doc.add_table(rows=1, cols=2)
        table.style = 'Table Grid'
        hdr = table.rows[0].cells
        hdr[0].text = 'Term (ENG)'
        hdr[1].text = 'Translation (UKR)'

        for t in unique_terms:
            row = table.add_row().cells
            row[0].text = t
            row[1].text = get_translation(t)

        new_doc.add_heading('Переклад тексту (UKR)', level=1)
        new_doc.add_paragraph(get_translation(text_block))

        new_doc.save(filename)
        auto_download_js(filename)
        lesson_num += 1

uploader = widgets.FileUpload(accept='.docx', multiple=False)
button = widgets.Button(description="Згенерувати уроки (A4)", button_style='success')
button.on_click(lambda b: process_full_book(uploader.value[0]['content'] if isinstance(uploader.value, list) else next(iter(uploader.value.values()))['content']))
display(uploader, button)
```
